```python
import pandas as pd
from autogluon.tabular import TabularDataset, TabularPredictor
```

```python
train_df = pd.read_csv('../data/CalorieConsumption/train.csv')
test_df = pd.read_csv('../data/CalorieConsumption/test.csv')
```

```python
train_df
```

```python
train = TabularDataset(train_df.drop(['ID'], axis=1))
test = TabularDataset(test_df.drop(['ID'], axis=1))
```

```python
predictor = TabularPredictor(label='Calories_Burned', eval_metric='root_mean_squared_error',).fit(train)
```

```python
predictor.leaderboard(train, silent=True)
```

```python
predictor.feature_importance(train)
```

```python
pred_y = predictor.predict(test)
```

```python
submission = pd.read_csv('../data/CalorieConsumption/sample_submission.csv')
submission['Calories_Burned'] = pred_y
```

```python
submission.to_csv('submission01.csv', index=False)
```
